```python
import numpy as np
from keras.preprocessing.text import Tokenizer
from numpy import array
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Embedding
```

```python
data="""
In the midst of a lush forest, sunlight filters through the canopy, casting dancing shadows on the ground below. Birds chirp happily, their melodies blending with the gentle rustling of leaves in the breeze. The air is filled with the sweet scent of blooming flowers, and a sense of tranquility envelopes the surroundings.

Amidst this natural splendor, a small wooden cabin stands, tucked away from the world. Smoke lazily rises from its chimney, signifying the presence of a warm hearth inside. Inside, a crackling fire provides a cozy ambiance, inviting anyone who stumbles upon it to seek solace from the outside world.

The cabin's walls are adorned with the artwork of its occupant, reflecting their love for nature and its wonders. A worn-out bookshelf stands tall, housing countless adventures and stories from far-off lands. A comfortable armchair sits beside the shelf, well-worn and inviting, offering a perfect spot to get lost in the pages of those tales.

A small kitchenette occupies one corner of the cabin, where the delicious aroma of freshly baked bread and simmering stew fills the air. A quaint wooden dining table stands nearby, with a vase of wildflowers as its centerpiece.

Beyond the cabin, a winding path leads deeper into the forest, teasing with the promise of undiscovered treasures. As the sun sets, painting the sky in hues of pink and orange, the forest comes alive with the songs of night creatures.

This serene hideaway, nestled in the heart of nature, offers a sanctuary for those seeking respite from the chaos of the world. A place where time slows down, and the connection with nature rekindles the spirit, reminding us of the beauty and wonder that exists all around us."

"""
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([data])
encoded_data = tokenizer.texts_to_sequences([data])[0]
encoded_data
```

```python
vocab_size = len(tokenizer.word_index)+1
print('vocabulary size: %d' % vocab_size)
```

```python
# we will create sequences of words to fit the model with one word as input and one as output
sequences = list()
for i in range(1,len(encoded_data)):
  sequence = encoded_data[i-1:i+1]
  sequences.append(sequence)
```

```python
sequences
```

```python
sequences = array(sequences)
x,y = sequences[:,0],sequences[:,1]
```

```python
y = to_categorical(y,num_classes = vocab_size)
y[:5 ]
```

```python
model = Sequential()
model.add(Embedding(vocab_size,10,input_length =1))
model.add(LSTM(50,return_sequences=False))
model.add(Dense(vocab_size,activation = 'softmax'))
model.summary
```

```python
model.summary()
```

```python
model.compile(loss = 'categorical_crossentropy',optimizer='adam',metrics = ['accuracy'])
```

```python
model.fit(x,y,epochs = 10)
```

```python
def generate_seq(model,tokenizer,enter_text,n_pred):
  in_text,result = enter_text,enter_text
  for _ in range(n_pred):
    encoded = tokenizer.texts_to_sequences([in_text])[0]
    encoded = array(encoded)

    yhat = model.predict_classes(encoded)

    out_word = ''
    for word,index in tokenizer.word_index.item():
      if index == yhat:
        out_word = word
        break
    in_text,result = out_word,result + ''+out_word
  return result
```

```python
print(generate_seq(model,tokenizer,'   ',2))
```
